**1. Up sampling**

## **NOTE : Data generated by upsampling is nothing but duplication of existing data**

```python
np.random.seed(123)

n_sample=1000
class_0_ratio=0.90
n_class_0=int(n_sample*class_0_ratio)
n_class_1=n_sample-n_class_0
```

```python
n_class_0
```

```python
n_class_1
```

```python
class_0 = pd.DataFrame({
    'feature_1' : np.random.normal(loc=0 , scale=1 , size=n_class_0),
    'feature_2' : np.random.normal(loc=0 , scale=1 , size=n_class_0),
    'target' : [0] * n_class_0
})

class_0
```

```python
class_1= pd.DataFrame({
    'feature_1' : np.random.normal(loc=0 , scale=1 , size=n_class_1),
    'feature_2' : np.random.normal(loc=0 , scale=1 , size=n_class_1),
    "target" : [1]*n_class_1
})
```

```python
sns.displot(class_1['feature_1'] , kde=True)
```

```python
df=pd.concat([class_0 , class_1])
```

```python
df
```

```python
df['target'].value_counts()
```

```python
df_minority=df[df['target'] == 1]
df_majority=df[df['target'] == 0]
```

```python
df_minority
```

```python
df_majority
```

```python
from sklearn.utils import resample

df_minority_upsampled=resample(df_minority , replace=True , n_samples=len(df_majority) , random_state=42)
df_minority_upsampled
```

```python
df_minority_upsampled.shape
```

```python
df_upsampled=pd.concat([df_majority , df_minority_upsampled ])
```

```python
df_upsampled
```

```python
df_upsampled.shape
```

**2 . Down Sampling**

```python
n_sample=1000
class_0_ratio=0.90
n_class_0=int(n_sample*class_0_ratio)
n_class_1=n_sample-n_class_0
```

```python
n_class_0
```

```python
n_class_1
```

```python
class_0=pd.DataFrame({
    'feature_1' : np.random.normal(loc=0 , scale=1 , size=n_class_0),
    'feature_2' : np.random.normal(loc=0 , scale=1 , size=n_class_0) , 
    'target' : [0] * n_class_0
})
```

```python
class_1=pd.DataFrame({
    'feature_1' : np.random.normal(loc=0 , scale=1 , size=n_class_1),
    'feature_2' : np.random.normal(loc=0 , scale=1 , size=n_class_1) , 
    'target' : [1] * n_class_1
})
```

```python
class_0.shape
```

```python
class_1.shape
```

```python
df=pd.concat([class_0 , class_1]).reset_index(drop=True)
```

```python
df
```

```python
df['target'].value_counts()
```

```python
from sklearn.utils import resample

df_majority=df[df['target']==0]
df_minority=df[df['target']==1]
```

```python
df_majority.shape
```

```python
df_minority.shape
```

```python
df_majority_downsample=resample(df_majority , replace=False,
                               n_samples=len(df_minority) ,
                               random_state=42)
```

```python
df_majority_downsample.shape
```

```python
df_minority.shape
```

```python
df_downsampled=pd.concat([df_majority_downsample  , df_minority]).reset_index(drop=True)
```

```python
df_downsampled.shape
```
