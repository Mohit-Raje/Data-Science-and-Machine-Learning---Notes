### One hot encoding : 
**One hot encoding**, also known as nominal encoding, is a technique used to represent categorical data as numerical data, which is more suitable for machine learning algorithms. In this technique, each category is represented as a binary vector where each bit corresponds to a unique category. For example, if we have a categorical variable "color" with three possible values (red, green, blue), we can represent it using one hot encoding as follows:

Red: [1, 0, 0]
Green: [0, 1, 0]
Blue: [0, 0, 1]


```python
## Create a simple dataframe 
df = pd.DataFrame({
    'color': ['red', 'blue', 'green', 'green', 'red', 'blue']
})
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder

onehot=OneHotEncoder()
encoded=onehot.fit_transform(df[['color']]).toarray()
```

```python
encoded
```

```python
encoded_features=pd.DataFrame(encoded , columns=onehot.get_feature_names_out())
```

```python
encoded_features
```

```python
final_df=pd.concat([df , encoded_features] , axis=1)
final_df
```

### Internal assignment

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
categorical_data=[]
for i in df.columns:
    if df[i].dtypes == 'category':
        categorical_data.append(i)
print(categorical_data)
```

```python
df[categorical_data]
```

```python
from sklearn.preprocessing import OneHotEncoder

onehotencoder=OneHotEncoder()
encoded_data=onehotencoder.fit_transform(df[categorical_data]).toarray()
```

```python
encoded_df=pd.DataFrame(data=encoded_data , columns=onehotencoder.get_feature_names_out())
encoded_df
```

```python
df=df.drop(categorical_data , axis=1)
```

```python
final_df=pd.concat([df , encoded_df] , axis=1)
```
