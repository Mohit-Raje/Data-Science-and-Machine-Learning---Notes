### Q4. Explain the concept of gradient descent. How is it used in machine learning?

Gradient descent is the optimization function which is responsible for minimizing the loss function by mainpulation the pararmeter.

In machine learning Gradient descent is used ti reduce the loss function by adjusting the parametres like : m and c in case of simple linear regression and m1,m2,m3...mn and c in case of multiple regression

A graph is plotted as follows : parameters(m,c)  VS Cost function(say mean squared error)
A Gradient descent curve is plotted where is the objective is to minimize the Cost function by converging toward the global minima (point at whhich the parameters are adjusted such that the cost function is minimum).
Derivative of cost function is taken wrt the parameters and multiplied with learning rate (learning rate decides the speed of convergence) and further subtracted from theta_j(parameter)

This process of convergence is repeated untill the global minima is reached.

### Q6. Explain the concept of multicollinearity in multiple linear regression. How can you detect and address this issue?

**Multicollinearity** occurs when two or more independent variables in a multiple linear regression model are highly correlated with each other.
This makes it difficult for the model to determine the individual effect of each predictor on the dependent variable, leading to unstable and unreliable coefficient estimates.


#### How to detect :**

How to Detect Multicollinearity?

1. Variance Inflation Factor (VIF)

Measures how much the variance of a regression coefficient is inflated due to multicollinearity.
Formula:
VIF = 1/(1-R^2)
 
Rule of Thumb:
VIF < 5 ‚Üí No multicollinearity
VIF between 5-10 ‚Üí Moderate multicollinearity
VIF > 10 ‚Üí High multicollinearity (problematic)


2. Correlation heatmap




#### How to address:**

How to Address Multicollinearity?

1. Remove One of the Highly Correlated Variables
If two variables are strongly correlated (e.g., Height and Weight in predicting BMI), consider keeping only one.

2. Use Principal Component Analysis (PCA)
PCA transforms correlated features into a new set of uncorrelated principal components, reducing redundancy.

3. Ridge Regression (L2 Regularization)
Unlike ordinary least squares (OLS), Ridge Regression shrinks correlated feature coefficients, reducing the impact of multicollinearity.

4. Collect More Data
