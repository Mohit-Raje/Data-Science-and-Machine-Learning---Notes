### Q2. What is the main advantage of using Lasso Regression in feature selection?

**Automatic Feature Selection:**

Lasso inherently selects the most important features by shrinking the coefficients of unimportant features to zero, reducing model complexity.

**Improves Interpretability:**

selecting only the most relevant features, Lasso makes the model simpler and easier to interpret.

**Handles High-Dimensional Data:** 

Useful when dealing with datasets with a large number of features, as it helps in reducing dimensionality and computational cost.

**Prevents Overfitting:** 

The regularization effect helps in avoiding overfitting by discouraging overly complex models.

### Q3. How do you interpret the coefficients of a Lasso Regression model?

**Steps to Interpret Lasso Coefficients:**

Non-Zero Coefficients: Features with non-zero coefficients are considered important for predicting the target variable. The larger the absolute value of the coefficient, the stronger the influence of that feature.

**Zero Coefficients:** 

Features with zero coefficients are considered irrelevant and are completely removed from the model, making Lasso useful for feature selection.

**Sign of Coefficients:**

A positive coefficient means that as the feature value increases, the target variable tends to increase.

A negative coefficient means that as the feature value increases, the target variable tends to decrease.

**Effect of Regularization Strength (α):**

Higher α (stronger regularization) → More coefficients shrink to zero, leading to fewer selected features.

Lower α (weaker regularization) → More features retain non-zero coefficients, behaving more like ordinary linear regression.

### Q4 . What are the tuning parameters that can be adjusted in Lasso Regression, and how do they affect the model's performance?

The primary tuning parameter in Lasso Regression is the regularization strength (α), but other hyperparameters can also be adjusted to optimize the model.
1. Regularization Parameter (α)
Effect: Controls the amount of shrinkage applied to the coefficients.


Impact on Model:


Higher α (stronger regularization):


More coefficients shrink to zero → Fewer selected features (stronger feature selection).


Reduces overfitting but may cause underfitting.


Lower α (weaker regularization):


More coefficients retain nonzero values → More features are kept in the model.


Higher risk of overfitting.


How to Tune: Use cross-validation (GridSearchCV or RandomizedSearchCV in Scikit-Learn) to find the optimal α value.


2. Maximum Iterations (max_iter)
Effect: Defines the number of iterations for optimization.


Impact on Model:


If too low, the model may not converge properly.


If too high, it may increase computation time unnecessarily.


3. Tolerance (tol)
Effect: Specifies the stopping criteria for optimization.


Impact on Model:


Higher tolerance: Faster convergence but lower accuracy.


Lower tolerance: More precise solution but longer training time.


4. Solver (selection)
Effect: Determines the optimization algorithm used for computing coefficients.


Common choices:


‘cyclic’ (default): Updates coefficients one at a time in a cycle.


‘random’: Updates coefficients in a random order (can be faster for large datasets).
