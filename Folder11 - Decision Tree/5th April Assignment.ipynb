### Q4. Use a decision tree algorithm, such as ID3 or C4.5, to train a decision tree model on the training set. Use cross-validation to optimize the hyperparameters and avoid overfitting.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix , classification_report
```

```python
param_grid = {
    'criterion': ['gini', 'entropy'],   # 'entropy' = ID3, 'gini' = CART (closest to C4.5)
    'max_depth': [3, 5, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
model = DecisionTreeClassifier(random_state=42)

grid_searchcv=GridSearchCV(estimator=model , param_grid=param_grid , cv=5 , scoring='accuracy')
```

```python
grid_searchcv.fit(X_train , y_train)
```

```python
grid_searchcv.best_params_
```

### Q7. Validate the decision tree model by applying it to new data or testing its robustness to changes in the dataset or the environment. Use sensitivity analysis and scenario testing to explore the uncertainty and risks.

```python
def predict_diabetes(model):
    print("Enter patient details below:")

    inputs = {}
    inputs['Pregnancies'] = int(input("Pregnancies: "))
    inputs['Glucose'] = float(input("Glucose: "))
    inputs['BloodPressure'] = float(input("Blood Pressure: "))
    inputs['SkinThickness'] = float(input("Skin Thickness: "))
    inputs['Insulin'] = float(input("Insulin: "))
    inputs['BMI'] = float(input("BMI: "))
    inputs['DiabetesPedigreeFunction'] = float(input("Diabetes Pedigree Function: "))
    inputs['Age'] = int(input("Age: "))

    # Convert to DataFrame
    input_df = pd.DataFrame([inputs])

    # Make prediction
    prediction = model.predict(input_df)[0]

    # Show result
    print("\n Prediction:")
    if prediction == 1:
        print(" The person is **Diabetic**.")
    else:
        print(" The person is **Not Diabetic**.")

# Call this with your trained model
predict_diabetes(dt)
```
