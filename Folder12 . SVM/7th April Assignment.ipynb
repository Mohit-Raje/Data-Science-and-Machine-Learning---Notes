```python
from sklearn import datasets
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Effects of kernel, C, epsilon, and gamma in SVR
üîπ 1. Kernel Function
What it does:
Transforms the input data into a higher-dimensional space where a linear regression can be applied.


Kernel Type	When to Use
linear	Data has a linear trend. Simple and efficient.
poly	Data shows polynomial (curved) patterns.
rbf	Default choice. Handles complex, nonlinear data well.
sigmoid	Rarely used. Similar to a neural net activation.
üß† Example:

If your dataset is a straight line (like height vs. weight), use linear.

If your data curves like a parabola, use poly.

If your data looks like a cloud with no obvious pattern, use rbf.

üîπ 2. C (Penalty / Regularization Term)
What it does:
Controls the trade-off between achieving a flat regression curve and allowing some points to be off by more than epsilon.


Value of C	Model Behavior
Low C	Allows wider margin ‚Üí more tolerance to error ‚Üí simpler model (can underfit).
High C	Narrow margin ‚Üí tries to fit every point ‚Üí complex model (can overfit).
üß† Example:

For noisy data (e.g., predicting stock prices), use low C.

For precise data (e.g., chemical measurements), use high C.

üîπ 3. Epsilon (Œµ)
What it does:
Creates an Œµ-tube around the regression line where no penalty is given for errors (i.e., predictions within this margin are ‚Äúgood enough‚Äù).


Value of Œµ	Model Behavior
Low Œµ	Narrow tube ‚Üí More points lie outside ‚Üí more support vectors ‚Üí higher model complexity.
High Œµ	Wide tube ‚Üí More tolerance ‚Üí simpler model ‚Üí fewer support vectors.
üß† Example:

For applications where small errors matter (e.g., heart rate prediction), use low Œµ.

For tolerant systems (e.g., forecasting temperature in 5¬∞C ranges), use high Œµ.

üîπ 4. Gamma (only for rbf, poly, sigmoid kernels)
What it does:
Controls the influence of each data point. It defines how far the impact of a single training point reaches.


Value of Œ≥ (gamma)	Model Behavior
Low Œ≥	Data points have a large influence ‚Üí smoother curve.
High Œ≥	Data points have narrow influence ‚Üí complex curve, may overfit.
üß† Example:

For generalized models (less sensitive to noise), use low gamma.

For detailed fitting (e.g., when every point matters), use high gamma.

```python
from sklearn import datasets
```

```python
data=datasets.load_breast_cancer()
```

```python
data.keys()
```

```python
X=pd.DataFrame(data=data['data'] , columns=data['feature_names'])
y=pd.DataFrame(data=data['target'] , columns=['Cancer'])
```

```python
X.head()
```

```python
X.shape
```

```python
y.head()
```

```python
y.value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.model_selection import train_test_split

X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.3 , random_state=42)
```

```python
std=StandardScaler()
X_train=std.fit_transform(X_train)
```

```python
X_test=std.transform(X_test)
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC()
svc.fit(X_train , y_train)
```

```python
y_pred=svc.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix , classification_report , precision_score , recall_score , accuracy_score
```

```python
print(confusion_matrix(y_test , y_pred))
print(classification_report(y_test , y_pred))
print(precision_score(y_test , y_pred))
print(recall_score(y_test , y_pred))
print(accuracy_score(y_test , y_pred))
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'C': [0.1, 1, 10, 100],               # Regularization parameter
    'kernel': ['linear', 'rbf', 'poly', 'sigmoid'],  # Kernel types
    'gamma': ['scale', 'auto', 0.01, 0.001],         # Kernel coefficient
    'degree': [2, 3, 4, 5]                # Used only for 'poly' kernel
}

gscv=GridSearchCV(SVC() , param_grid , cv=5 , refit=True , verbose=1)
gscv
```

```python
gscv.fit(X_train , y_train)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```

```python
svc1=SVC(C=1 , degree=2 , gamma='scale' , kernel='linear'  , probability=True)
```

```python
svc1.fit(X_train , y_train)
```

```python
y_pred1=svc1.predict(X_test)
```

```python
print(confusion_matrix(y_test , y_pred1))
print(classification_report(y_test , y_pred1))
print(precision_score(y_test , y_pred1))
print(recall_score(y_test , y_pred1))
print(accuracy_score(y_test , y_pred1))
```

```python
from sklearn.metrics import roc_curve , roc_auc_score , f1_score 
```

```python
y_proba=svc1.predict_proba(X_test)[: , 1]
```

```python
fpr , tpr , threshold = roc_curve(y_test , y_proba)
```
