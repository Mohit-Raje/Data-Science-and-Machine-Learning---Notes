## Train - Test split

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.iloc[: , :-1]
y=df.iloc[: , -1]
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.30 ,random_state=42)
```

## Pipleline

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
```

```python
columns=list(X_train.columns)
cat_features=[col for col in columns if df[col].dtypes=='object']
numeric_features = [col for col in columns if df[col].dtypes!='object']
print(cat_features)
print()
print(numeric_features)
```

```python
cat_pipeline=Pipeline(
    
    steps=[
        ('imputer' , SimpleImputer(strategy='most_frequent')),
        ('ohe' , OneHotEncoder()),
        ('scaler' , StandardScaler(with_mean=False))
    ]    

)
```

```python
numeric_pipeline=Pipeline(
    
    steps=[
        ('imputer' , SimpleImputer(strategy='mean')),
        ('scaler' , StandardScaler())
    ]
    
)
```

```python
preprocessor=ColumnTransformer([
    ('cat_pipeline' , cat_pipeline , cat_features) ,
    ('numeric_pipeline' , numeric_pipeline , numeric_features)
])
```

```python
preprocessor
```

```python
X_train_transformed=preprocessor.fit_transform(X_train)

X_train_dense=X_train_transformed.toarray()

X_train_transformed=X_train_dense

X_train_transformed
```

```python
X_test_transformed=preprocessor.transform(X_test)
X_test_transformed=X_test_transformed.toarray()
X_test_transformed
```

## Selecting the best algorithm

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB , BernoulliNB 
```

```python
models={
    'Random Forest': RandomForestClassifier(),
    'Decision Tree' : DecisionTreeClassifier(),
    'SVC' : SVC() ,
    'Logistic Regression' : LogisticRegression() ,
    'Gaussain Naive Bayes' : GaussianNB() ,
    'Bernoulli Naive Bayes' : BernoulliNB()
}
```

```python
from sklearn.metrics import accuracy_score
```

```python
list(models.values())[0]
```

```python
list(models.keys())[0]
```

```python
def get_best_accuracy(X_train , X_test , y_train,  y_test, models):
    
    report={}
    for i in range(len(models)):
        model=list(models.values())[i]
        model.fit(X_train , y_train)
        
        y_pred=model.predict(X_test)
        accuracy=accuracy_score(y_test  , y_pred)
        
        report[list(models.keys())[i]]=accuracy
    return report
```

```python
report=get_best_accuracy(X_train_transformed , X_test_transformed , y_train,  y_test, models)
report
```
