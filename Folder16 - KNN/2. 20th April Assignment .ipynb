## Q1. What is the KNN algorithm?

Q1. What is the KNN algorithm?
KNN (K-Nearest Neighbors) is a supervised machine learning algorithm used for both classification and regression tasks.

üß† How it Works (Classification):
When given a new data point to classify, KNN:

Looks at the ‚Äòk‚Äô closest data points from the training set.

Counts the classes of those neighbors.

Assigns the most common class among them to the new point.

üì¶ Example:
Suppose k=3 and a new point is surrounded by:

2 "Cats"

1 "Dog"

Then, the KNN algorithm will classify the point as a "Cat".

üìè Key Ideas:
It uses distance (commonly Euclidean) to find neighbors.

Lazy learner: It doesn't learn a model; it just stores the data.

Works best when data is not too high-dimensional and is scaled properly.

üßÆ Distance Used:
Euclidean (default), Manhattan, Minkowski, etc.

You can customize the distance function in the model.

Pros:
Simple to understand and implement

No training time

Naturally handles multi-class problems

Cons:
Slow prediction on large datasets

Sensitive to irrelevant features and scaling

Doesn't work well with high-dimensional data (curse of dimensionality)

## Q3. What is the difference between KNN classifier and KNN regressor?

The main difference KNN classifier and regressor is the way in which they combine results obtained by the n neighbours . In case of classification majority voting is used i.e the class to which majority of the neighbour belongs that calss is assigned to the test data point . IN case of regression n nearest neighbours of the test data are considered and their avg is the result .

KNN Classifier uses majority voting among neighbors to predict the class, while KNN Regressor uses the average (or weighted average) of neighbors' values to predict a continuous output

Both methods use the same distance-based approach to find neighbors, but differ in how the final prediction is made.

## Q5. What is the curse of dimensionality in KNN?

The curse of dimensionality refers to the problems that arise when working with data in high-dimensional spaces. In KNN, as the number of features increases, the distance between points becomes less meaningful ‚Äî all points start to appear equally far. This degrades the performance of KNN, as it relies on meaningful distance to find nearest neighbors. It often leads to high variance and poor generalization.

## Q6. How do you handle missing values in KNN?

In KNN, missing values are typically handled through imputation. Common methods include replacing missing values with the mean, median, or mode of the feature. More advanced methods use KNN imputation, where the missing value is filled based on the values from the nearest neighbors.

## Q9. What is the difference between Euclidean distance and Manhattan distance in KNN?

Euclidean distance is the straight-line distance between two points (like using a ruler).
Manhattan distance is the sum of the absolute differences along each dimension (like moving in a grid or city blocks).



Euclidean is better when all features are continuous and there's no strong axis alignment.

Manhattan is often better when data is sparse, or has many dimensions, or when features are aligned with axes.



Euclidean distance is like drawing a straight line between two points ‚Äî it's best when the data is continuous and features aren't restricted to any particular direction.

Manhattan distance is like walking in a city ‚Äî you move along the blocks, not diagonally. It's better when the data is sparse, has lots of zeros, or is high-dimensional, because it handles those cases more reliably.

## Q10. What is the role of feature scaling in KNN?


Feature scaling is crucial in KNN because the algorithm relies on distance calculations. If features are on different scales (e.g., age in years vs. income in lakhs), the feature with the larger scale will dominate the distance ‚Äî and this can lead to biased or incorrect predictions.
