## Classifier

```python
from sklearn.datasets import load_iris
```

```python
X  , y = load_iris(return_X_y=True)
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X, y, test_size=0.33 ,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_clf=KNeighborsClassifier()
```

```python
knn_clf.fit(X_train  , y_train)
```

```python
y_pred=knn_clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix , classification_report , precision_score , recall_score
```

```python
print(accuracy_score(y_test , y_pred))
print(confusion_matrix(y_test , y_pred))
print(classification_report(y_test , y_pred))
print(precision_score(y_test , y_pred , average='weighted'))
print(recall_score(y_test , y_pred , average='weighted'))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={

    'n_neighbors' : [1,2,3,4,5,6,7,8,9],
    'algorithm' : ['auto' , 'ball_tree' , 'kd_tree' , 'brute'],
    'p' :[1,2]
}
```

```python
knn_clf1=KNeighborsClassifier()
```

```python
gscv=GridSearchCV(knn_clf1 , param_grid=param_grid , refit=True , cv=3 , verbose=2)
```

```python
gscv.fit(X_train , y_train)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```

```python
import numpy as np
```

```python
np.square(8-4)
```

```python
def euclidean_distance(x1 , y1 , x2 , y2):
    x=np.square(x1-x2)
    y=np.square(y1-y2)
    z=x+y
    return round(np.sqrt(z) , 2)
```

```python
print(euclidean_distance(5 , 4 , 7 , 8))
```

```python
def manhattan_dist(x1 , y1 , x2 , y2):
    x=np.abs(x1-x2)
    y=np.abs(y1-y2)
    z=x+y
    return z
```

```python
print(manhattan_dist(5 , 4 , 7 , 8))
```

## Regression

```python
import pandas as pd
```

```python
df=pd.read_csv("Boston.csv")
```

```python
df.head()
```

```python
X=df.iloc[: , :-1]
y=df.iloc[: , -1]
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.33 ,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn_reg1=KNeighborsRegressor()
```

```python
knn_reg1.fit(X_train , y_train)
```

```python
y_pred=knn_reg1.predict(X_test)
```

```python
from sklearn.metrics import r2_score , mean_absolute_error, mean_squared_error
```

```python
print(r2_score(y_test , y_pred))
print(mean_absolute_error(y_test , y_pred))
print(mean_squared_error(y_test , y_pred))
```

```python
knn_reg2=KNeighborsRegressor(weights='distance')
```

```python
knn_reg2.fit(X_train , y_train)
```

```python
y_pred=knn_reg2.predict(X_test)
```

```python
print(r2_score(y_test , y_pred))
print(mean_absolute_error(y_test , y_pred))
print(mean_squared_error(y_test , y_pred))
```

```python
param_grid={
    'n_neighbors' : [1,2,3,4,5,6,7,8,9],
    'algorithm' : ['auto' , 'ball_tree' , 'kd_tree' , 'brute'],
    'p' :[1,2] ,
    'weights' : ['uniform' , 'distance']
}
```

```python
gscv=GridSearchCV(KNeighborsRegressor() , param_grid=param_grid , cv=3 , refit=True , verbose=2)
```

```python
gscv.fit(X_train , y_train)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```
