```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("healthcare.csv")
```

```python
df.head()
```

```python
plt.scatter(df.iloc[: , 0] , df.iloc[: , 1])
```

```python
from sklearn.ensemble import IsolationForest
```

```python
isolation_tree=IsolationForest(contamination=0.2)
```

```python
isolation_tree
```

```python
isolation_tree.fit(df)
```

```python
prediction=isolation_tree.predict(df)
```

```python
prediction
```

```python
plt.scatter(df.iloc[: , 0] , df.iloc[: , 1] , c=prediction)
```

```python
set(prediction)
```

```python
index=np.where(prediction < 0)
index
```
